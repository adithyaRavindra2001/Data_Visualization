```python
import pandas as pd
import numpy as np
```

```python
import cufflinks as cf
```

```python
import plotly.offline as plyo
```

```python
plyo.init_notebook_mode(connected=True)
```

```python
a=np.random.standard_normal((250,5)).cumsum(axis=0)
```

```python
index=pd.date_range('2019-1-1',freq='B',periods=len(a))
```

```python
df=pd.DataFrame(100+5*a,columns=list('abcde'),index=index)
```

```python
df.head()
```

```python
plyo.iplot(df.iplot(asFigure=True),filename='ply_01')
```

```python
plyo.iplot(
         df[['a', 'b']].iplot(asFigure=True,
         theme='polar', 
         title='A Time Series Plot', 
         xTitle='date', 
         yTitle='value', 
         mode={'a': 'markers', 'b': 'lines+markers'}, 
         symbol={'a': 'circle', 'b': 'diamond'}, 
         size=3.5, 
         colors={'a': 'blue', 'b': 'magenta'}, 
         ),
         # image='png',
 filename='ply_02'
 )
```

```python
plyo.iplot(
df.iplot(kind='hist',subplots=True,bins=15,asFigure=True),
filename='ply_03')
```

```python
raw=pd.read_csv('fxcm_eur_usd_eod_data.csv',parse_dates=True)
raw.head()
```

```python
raw=pd.DataFrame(raw[['OpenAsk', 'HighAsk', 'LowAsk', 'CloseAsk']])
```

```python
raw
```

```python
qf=cf.QuantFig(raw,
              title='EUR/USD Exchange Rate',
              legend='top',
              name='EUR/USD')
```

```python
plyo.iplot(qf.iplot(asFigure=True),
          filename='qf_01')
```

```python
qf.add_bollinger_bands(periods=15,boll_std=2)
```

```python
plyo.iplot(qf.iplot(asFigure=True),
          filename='qf_02')
```
